```python
import requests
import json

# Taking user input to show weather details of any city
city_name = input('Enter the city name: ')
api_key = '68a7cd6a78bb43b7803e9bab558b0a88'
url2 = f'https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}'

response2 = requests.get(url2)
data2 = response2.json()

if response2.status_code == 200:
    print('Weather:', data2['weather'][0]['main'])
    print('City:', data2['name'])
    print('Temperature:', round(data2['main']['temp'] - 273.15), '°C')
else:
    print("Enter the correct city please")
```
